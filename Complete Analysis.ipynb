# Sellers Analize

```python
df = pd.merge(left = sellers, 
              right= items,
              how = 'left',
             on = 'seller_id')
df.head()
```

```python
cols = ['seller_city']
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
```

```python
df['Total_sales'] = (df['price'].fillna(0) + df['freight_value'].fillna(0))*(df['order_item_id'].fillna(0))
```

```python
df.groupby('seller_id').sum()['Total_sales']

top_sellers = pd.merge(left = pd.DataFrame(df.groupby('seller_id').sum()['Total_sales']), 
         right =df.drop('Total_sales', axis =1),
         left_index= True,
         right_on='seller_id'       
        )[['Total_sales','seller_id','seller_city','seller_state']]
top_sellers.drop_duplicates(inplace = True)
top_sellers.sort_values(['Total_sales'], ascending = False, inplace = True)
top_sellers.head(10)
```

# Buyers Analize

```python
df2 = pd.merge(right = orders,
               left = items , 
               on= 'order_id',
               how = 'left')

df3 = pd.merge(right = df2, 
               left = customers, 
               on = 'customer_id',
               how = 'left')
df3['Total_purchases'] = df3['order_item_id'].fillna(0)*(df3['price'].fillna(0)+ df3['freight_value'].fillna(0))
```

```python
cols = ['customer_city']
df3[cols] = df3[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
```

```python
df3.groupby('customer_id').sum()['Total_purchases']

top_customers = pd.merge(left = pd.DataFrame(df3.groupby('customer_id').sum()['Total_purchases']), 
         right =df3.drop(['Total_purchases'], axis = 1),
         left_index= True,
         right_on='customer_id'       
         )[['Total_purchases','customer_id','customer_city','customer_state']]
top_customers.drop_duplicates(inplace = True)
top_customers.sort_values(['Total_purchases'], ascending = False, inplace = True)
top_customers.head(10)
```

## Preparatives

```python
import folium
```

```python
geolocation = pd.read_csv('olist_geolocation_dataset.csv')

geolocation.sort_values(by = 'geolocation_zip_code_prefix')

# fixes special characters
# deletes duplicates from zip code lat/long

geolocation[['geolocation_city']] = geolocation[['geolocation_city']].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
geolocation = geolocation.groupby(['geolocation_zip_code_prefix', 'geolocation_city','geolocation_state']).mean() 
geolocation.sort_values(by = 'geolocation_city', inplace=True)
geolocation.reset_index(level=['geolocation_city','geolocation_state'], inplace=True)
```

```python
def map_lat_long(df,MAX_RECORDS=50 ):
    
    m = folium.Map(
        location=[-16, -60],    # Centro do Brasil    
        zoom_start=4)
    
    for idx in range(min(MAX_RECORDS, df.shape[0])):
        folium.Marker([df[['geolocation_lat']].iloc[idx][0],
            df[['geolocation_lng']].iloc[idx][0]]).add_to(m)
    return(m)
```

## Top sellers location

```python
top_sellers_zip = pd.merge(top_sellers, df[['seller_id', 'seller_zip_code_prefix']], left_on = 'seller_id', right_on = 'seller_id')

top_sellers_map = pd.merge(top_sellers_zip,geolocation, left_on= 'seller_zip_code_prefix', right_index=True ).drop_duplicates().sort_values('Total_sales', ascending = False).reset_index(drop = True)

map_lat_long(top_sellers_map)
```

## Top Customers location

```python
top_customers_zip = pd.merge(top_customers, df3[['customer_id', 'customer_zip_code_prefix']], left_on = 'customer_id', right_on = 'customer_id')

top_customers_map = pd.merge(top_customers_zip,geolocation, left_on= 'customer_zip_code_prefix', right_index=True ).drop_duplicates().sort_values('Total_purchases', ascending = False).reset_index(drop = True)

map_lat_long(top_customers_map)
```

## Clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot   as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
basexxx = df_freight[['dimensoes', 'product_weight_kg', 'freight_value']]
basexxx = basexxx.dropna()

modelo = KMeans(n_clusters=6,   
                n_jobs = 4,                 
                max_iter = 300, 
                random_state= 1)
basexxx['cluster'] = modelo.fit_predict(basexxx)


#ordena nome dos clusters por tamanho/peso
y = basexxx.groupby('cluster').mean().sort_values('freight_value')
basexxx['cluster'] = basexxx['cluster']*100

for i in y.index:
    if i != 0:
        basexxx.loc[(basexxx['cluster'] == y.index[i]*100), 'cluster'] = i
basexxx.loc[(basexxx['cluster'] == y.index[0]*100), 'cluster'] = 0
basexxx.head()
```

```python
cols = ['product_weight_kg', 'dimensoes','freight_value']
for col in cols:
    sns.boxplot(y = col, x = 'cluster', data = basexxx)
    plt.show()
```

```python
cols = ['product_weight_kg', 'dimensoes','freight_value']
for col in cols:
    plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    sns.violinplot(x = 'cluster', y=col, data = basexxx, width=4)
    plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
fig.subplots_adjust(bottom=-0.15,top=1.2)
ax.scatter(
    basexxx['dimensoes'], 
    basexxx['product_weight_kg'], 
    basexxx['freight_value'],
    s = 10, c = basexxx['cluster'])
ax.set_xlabel('Dimensoes')
ax.set_ylabel('Peso')
ax.set_zlabel('Frete')
plt.show()
```

```python
cols = ['review_comment_message', 'review_comment_title']
avaliacoes[cols] = avaliacoes[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.lower())
avaliacoes.head()
```

```python
stopwords_txt = ['um','com', 'na', 'e', "da","para","comprei","pois", "mesmo" "que", "em", 
                 "estou", "o", "a", "de", "do", 'sao', 'dos', 'santo', 'santa', 'foi', 'produto',
                 'são', 'que', 'eu', 'ate', 'até']
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(stopwords_txt)

# Generate a word cloud image
wordcloud = WordCloud(width=2000, height=1080,stopwords=stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.figure( figsize=(20,10) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
avaliacoes['review_score'].hist( grid = False)
```

```python
avaliacoes_5 = avaliacoes[avaliacoes['review_score']==5]
avaliacoes_4 = avaliacoes[avaliacoes['review_score']==4]
avaliacoes_3 = avaliacoes[avaliacoes['review_score']<=3]

lista_de_bases = [avaliacoes_5, avaliacoes_4, avaliacoes_3]
```

```python

titulo = ["Nota de Satisfacao = 5 ","Nota de Satisfacao = 4 ","Nota de Satisfacao < 4 "]
for idx, base in enumerate(lista_de_bases):
    texto = base['review_comment_title'].dropna()
    text = " ".join(review for review in texto)
    lista = pd.DataFrame(text.split())
    lista.columns = ['palavras']
    lista = lista[~lista['palavras'].isin(stopwords_txt)]
    lista['flag']=1
    lista.loc[lista['palavras']=='otima', 'palavras']= 'otimo'
    lista.loc[lista['palavras']=='boa'  , 'palavras']= 'bom'
    lista.loc[lista['palavras']== 'recomendado'  , 'palavras']= 'recomendo'
    lista.loc[lista['palavras']==  'pessima' , 'palavras']= 'pessimo'
    lista.loc[lista['palavras']==  'entregue' , 'palavras']= 'entrega'
    display(lista['flag'].sum())

    plt.figure(figsize = [15, 4])
    plt.plot(lista.groupby('palavras').count().sort_values(by = 'flag', ascending = False).head(12), '.', markersize =15)
    plt.xticks(rotation=20)
    plt.ylabel("Volume de aparições unicas")
    plt.title(titulo[idx] )
    plt.show()
```
