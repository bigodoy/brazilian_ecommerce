# Unindo as bases

```python
clientes2 =pd.merge(
    clientes,localizacao, 
    how = 'left',
    left_on =  'customer_zip_code_prefix', 
    right_on = 'geolocation_zip_code_prefix'
)
clientes2.head()
```

```python
vendedores2 =pd.merge(
    vendedores,localizacao, 
    how = 'left',
    left_on =  'seller_zip_code_prefix', 
    right_on = 'geolocation_zip_code_prefix'
)
vendedores2.head()
```

```python
itens2 =pd.merge(
    itens ,vendedores , 
    how = 'left',
    on =  'seller_id'
)
# itens2.head()
```

```python
itens3 =pd.merge(
    itens2 ,produtos, 
    how = 'left',
    on =  'product_id'
)
itens3.head()
```

```python
encomendas2 =pd.merge(
    encomendas ,clientes, 
    how = 'left',
    on =  'customer_id'
)
encomendas2.head()
```

```python
encomendas3 =pd.merge(
    encomendas2 ,pagamentos, 
    how = 'left',
    on =  'order_id'
)
encomendas3.head()
```

```python
encomendas4 =pd.merge(
    encomendas3 ,itens3, 
    how = 'left',
    on =  'order_id'
)
encomendas4.head()
```

```python
df =pd.merge(
    encomendas4 ,avaliacoes, 
    how = 'left',
    on =  'order_id'
)
df.head()
```
